```python
data_train = pd.DataFrame(df2.Close[0: int(len(df2)*0.70)])
data_test = pd.DataFrame(df2.Close[ int(len(df2)*0.70) : len(df2)])
```

```python
data_train
```

```python
data_test
```

```python
print(data_train.shape)
print(data_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
```

```python
data_train_scale = scaler.fit_transform(data_train)
data_train_scale
```

```python
x = []
y = []

for i in range(100, len(data_train_scale)):
    x.append(data_train_scale[i-100:i])
    y.append(data_train_scale[i, 0])
```

```python
x, y = np.array(x), np.array(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
from keras.layers import Dense,Dropout,LSTM
from keras.models import Sequential

model = Sequential()
model.add(LSTM(
    units=50,
    activation='relu',
    return_sequences=True,
    input_shape=(x.shape[1], 1)
))

model.add(Dropout(0.2))

model.add(LSTM(
    units=60,
    activation='relu',
    return_sequences=True
))
model.add(Dropout(0.3))

model.add(LSTM(
    units=80,
    activation='relu',
    return_sequences = True
))

model.add(Dropout(0.4))

model.add(LSTM(
    units=120,
    activation='relu'
))

model.add(Dense(1))

model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)

model.fit(
    x,
    y,
    epochs =10,
    batch_size = 32,
    verbose = 1
)


    
```

```python
model.summary()
```

```python
Hundred_days = data_train.tail(100)
data_test = pd.concat([Hundred_days,data_test],ignore_index=True)
```

```python
data_test
```

```python
data_test_scale = scaler.fit_transform(data_test)
data_test_scale
```

```python
x = []
y = []

for i in range(100, data_test_scale.shape[0]):
    x.append(data_test_scale[i-100:i])
    y.append(data_test_scale[i,0])
```

```python
x, y = np.array(x), np.array(y)
```

```python
y_predict = model.predict(x)
```

```python
scaler.scale_
```

```python
y_predict
```

```python
scale = 1/scaler.scale_
```

```python
scale
```

```python
y_predict = y_predict*scale
```

```python
plt.figure(figsize = (12,8))
plt.plot(y,'b', label = 'orginal price',linewidth = 1 )
plt.plot(y_predict,'g',label = 'predicted price',linewidth =1)
plt.xlabel('Stock Price')
plt.ylabel('Time ')
plt.title('Stock Price Prediction')
plt.legend()
plt.show()
```
